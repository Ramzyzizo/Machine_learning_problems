```python
import sklearn
from sklearn.utils import shuffle
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
from sklearn import linear_model,preprocessing #convert to numerical value
```

```python
data = pd.read_csv("car.data")
print(data.head())
```

```python
le = preprocessing.LabelEncoder()
buying = le.fit_transform(list(data["buying"])) #get column to list in numerical values
maint = le.fit_transform(list(data["maint"])) 
door = le.fit_transform(list(data["door"])) 
persons = le.fit_transform(list(data["persons"])) 
lug_boot = le.fit_transform(list(data["lug_boot"])) 
safety = le.fit_transform(list(data["safety"])) 
cls = le.fit_transform(list(data["class"])) 


predict = "class"
```

```python
x = list(zip(buying,maint,door,persons,lug_boot,safety))
y = list(cls)

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.2)

model = KNeighborsClassifier(n_neighbors=9)
model.fit(x_train,y_train)
acc=model.score(x_test,y_test)
print(acc)
```

```python
predicted = model.predict(x_test)
name=["unacc","acc","good","vgood"]
for x in range(len(predicted)):
    print("Predicted: ",name[predicted[x]],"Data: ", x_test[x], "Actual: ",name[y_test[x]])
    n=model.kneighbors([x_test[x]],9,True)
    print("n:", n)
```
